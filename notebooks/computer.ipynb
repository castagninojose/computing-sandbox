```python
# plot de sucesiones de funciones
def f_n(x, n):
    return n*dom / (1+(n*dom)**2)

fig = go.Figure()
dom = pd.Series(np.linspace(0, 1, 100))
for n in range(1,31):
    img_n = f_n(dom, n)
    fig.add_trace(go.Scatter(x=dom,y=img_n, name=f'n:{n}'))
    

fig.update_layout(title=r'$f_n(x) :\frac{nx}{1+n^2x^2}$')    
iplot(fig)
```
